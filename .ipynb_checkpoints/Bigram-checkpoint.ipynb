```python
chars = sorted(list(set(''.join(words))))
stoj = {i:j for j,i in enumerate(chars)}
stoj['<S>'] = 26
stoj['<E>'] = 27
```

```python
for w in words:
    bucket = ['<S>'] + list(w) + ['<E>']
    for i in range(len(bucket)-1):
        jx1 = stoj[bucket[i]]
        jx2 = stoj[bucket[i+1]]
        N[jx1, jx2] += 1
```
