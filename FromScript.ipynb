```python
import pandas as pd
import numpy as np
```

```python
file = ("LewistonID.xlsx")
```

```python
df = pd.read_excel(file)
df.head(30)
```

```python
print(df.at[2396, "Descriptio"])
```

```python
temp = []
for i in df.index:
    temp.append(df.at[i, 'Descriptio'])
    if pd.isna(temp[i]):
        continue
    sL = temp[i].find("To")
    if sL == -1:
        sL = temp[i].find("to")
        df.at[i, 'To'] = temp[i][sL+2:]
    df.at[i, 'To'] = temp[i][sL+2:]

    print(i,df.at[i, "From"])
```
